# PS4: Restricted Boltzmann Machines and MNIST Digit Recall

Handwritten digit images from the [MNIST database](https://en.wikipedia.org/wiki/MNIST_database) contain rich spatial structure — stroke patterns, curves, and intersections that co-occur across examples of the same digit. In this problem set, we explore how two associative memory models — the classical Hopfield network and the restricted Boltzmann machine (RBM) — store and recall digit patterns from corrupted inputs.

The [Hopfield network](https://en.wikipedia.org/wiki/Hopfield_network) stores patterns in a Hebbian weight matrix and retrieves them by minimizing an energy function through iterative state updates. The [restricted Boltzmann machine](https://en.wikipedia.org/wiki/Restricted_Boltzmann_machine) takes a different approach: it learns a probability distribution over binary pixel patterns and reconstructs inputs via block Gibbs sampling. We compare both models on the same recall task — recovering a corrupted MNIST digit — and then examine how model capacity and training data affect reconstruction quality.

> Converts a flattened state vector $\mathbf{s}$ (with values in $\{-1, +1\}$ or $\{0, 1\}$) back to a 2D image matrix. The function reshapes the vector, transposes to match row-major ordering, and replaces $-1$ values with $0$ for grayscale display.

Let's implement the `decode(...)` function:

> __`decode(s; number_of_rows, number_of_columns)`__
>
> Converts a flattened state vector $\mathbf{s}$ (with values in $\{-1, +1\}$ or $\{0, 1\}$) back to a 2D image matrix. The function reshapes the vector, transposes to match row-major ordering, and replaces $-1$ values with $0$ for grayscale display.

Let's implement the `decode(...)` function:

```julia
function decode(s::Vector{<:Number}; number_of_rows::Int=28, number_of_columns::Int=28)::Array{<:Number,2} #decode function, s is vectors
    
    # initialize -
    X = reshape(s, number_of_rows, number_of_columns) |> X -> transpose(X) |> Matrix; #reshape into 28 x 28 matrix
    X̂ = replace(X, -1 => 0); #replace -1 with 0

    # return -
    return X̂
end;
```

> __`corrupt(v, frac; seed)`__
>
> Corrupts a binary $\{-1, +1\}$ pattern $\mathbf{v}$ by flipping a random fraction `frac` of its bits. Each flipped bit changes sign: $v_i \leftarrow -v_i$. This simulates noisy or partially observed inputs for the recall task.

Let's implement the `corrupt(...)` function:

```julia
function corrupt(v::Vector{Int64}, frac::Float64; seed::Int=42)::Vector{Int64} #corrupt function
    
    # initialize -
    n = length(v); #length of the vector
    number_of_flips = round(Int, frac * n); #number pixel flipped 
    rng = MersenneTwister(seed); #random number generator 
    flip_idx = StatsBase.sample(rng, 1:n, number_of_flips, replace=false); #random position flip
    
    # corrupt the pattern -
    v_corrupted = copy(v); #makes a copy to prevent changes
    v_corrupted[flip_idx] .*= -1; #multiply pixel by 1

    # return -
    return v_corrupted
end;
```

> __`tanimoto(a, b)`__
>
> Computes the Tanimoto (Jaccard) similarity between two $\{0, 1\}$ binary vectors. The Tanimoto coefficient is $T(\mathbf{a}, \mathbf{b}) = \frac{\langle \mathbf{a}, \mathbf{b} \rangle}{\|\mathbf{a}\|_1 + \|\mathbf{b}\|_1 - \langle \mathbf{a}, \mathbf{b} \rangle}$, which equals 0 when the patterns share no "on" pixels and 1 when they are identical. Unlike bit accuracy, Tanimoto is not inflated by background pixels.

Let's implement the `tanimoto(...)` function:

```julia
function tanimoto(a::Vector{<:Number}, b::Vector{<:Number})::Float64 #numeric vectors, returns a similarity score
    
    # initialize -
    ab = dot(a, b); #dot product
    denom = sum(a) + sum(b) - ab; #denmoniator calculation

    # return -
    return denom == 0 ? 0.0 : ab / denom #compute similarity 
end;
```

> __`to_binary(v)`__
>
> Converts a $\{-1, +1\}$ integer vector to $\{0, 1\}$ floating-point encoding via the mapping $v \mapsto (v + 1)/2$. This is used before computing Tanimoto similarity, which expects $\{0, 1\}$ inputs.

Let's implement the `to_binary(...)` function:

```julia
to_binary(v::Vector{<:Number})::Vector{Float64} = Float64.((v .+ 1) ./ 2); #convert to binary 
```

## Task 1: Memory Matrix and Hopfield Recall
A [Hopfield network](https://en.wikipedia.org/wiki/Hopfield_network) stores binary patterns as a symmetric weight matrix and retrieves them by iteratively updating each unit's state to minimize a global energy function. Given a set of $N$ stored patterns $\left\{\boldsymbol{\xi}^{(1)},\ldots,\boldsymbol{\xi}^{(N)}\right\}$ with $\boldsymbol{\xi}^{(\mu)}\in\{-1,+1\}^{n}$, the Hebbian learning rule constructs the weight matrix as an outer-product sum:

$$
\mathbf{W} = \frac{1}{N}\sum_{\mu=1}^{N}\boldsymbol{\xi}^{(\mu)}\left(\boldsymbol{\xi}^{(\mu)}\right)^{\top},\qquad w_{ii}=0.
$$

Recall dynamics update the state as $\mathbf{v}^{(t+1)} = \texttt{sign}\!\left(\mathbf{W}\,\mathbf{v}^{(t)}\right)$, which is guaranteed to converge to a local minimum of the energy function $E(\mathbf{v}) = -\frac{1}{2}\mathbf{v}^{\top}\mathbf{W}\,\mathbf{v}$ (where we assume the bias terms are zero). In this task, we build a memory matrix from MNIST digit examples and test whether Hopfield dynamics can recover a stored pattern from a corrupted input.

> __Memory matrix construction__
>
> We load `number_of_examples` images of `digit_for_experiment` from the MNIST dataset, threshold each pixel at $0.5$ to obtain a binary $\{0, 1\}$ image, convert to $\{-1, +1\}$ encoding, and store each flattened pattern as a column of $\mathbf{X}_{\pm 1} \in \mathbb{Z}^{n_{pixels} \times N}$.

Let's load the MNIST data and store the results in `digits_dict::Dict` (maps each digit label to its image array), `X::Array{Float64,2}` (the $\{0,1\}$ pattern matrix), and `X_pm1::Array{Int64,2}` (the $\{-1,+1\}$ pattern matrix used by the Hopfield network).

```julia
digits_dict, X, X_pm1 = let

    # initialize -
    digits_dict = MyMNISTHandwrittenDigitImageDataset(number_of_examples = number_of_examples + 10); # MNIST image dictionary (digit => 28x28xN array
    X = zeros(Float64, number_of_pixels, number_of_examples); # {0,1} encoding # memory matrix in {0,1} encoding (pixels × patterns)
    X_pm1 = zeros(Int64, number_of_pixels, number_of_examples); # {-1,+1} encoding # memory matrix in {-1,+1} encoding (pixels × patterns)

    # TODO: fill in X and X_pm1
    # for each i in 1:number_of_examples
    #   get the i-th image of digit_for_experiment from digits_dict
    #   flatten to a vector, threshold at 0.5 to get a binary {0,1} pattern
    #   convert to {-1,+1} and store in X_pm1[:, i]
    
 # load and binarize each image -
    for i in 1:number_of_examples
        img = digits_dict[digit_for_experiment][:, :, i]; # raw grayscale image (28 × 28)
        v = Float64.(img)[:];       # flatten to 784-element vector
        b = (v .> 0.5);             # threshold to binary {0,1}
        X[:, i] = Float64.(b);      # store {0,1} pattern
        X_pm1[:, i] = Int64.(2 .* b .- 1); # convert {0,1} → {-1,+1}
    end
    println("Memory matrix: $(size(X_pm1, 1)) pixels x $(size(X_pm1, 2)) stored patterns")

    # return -
    (digits_dict, X, X_pm1)
end;
```

> __Visualize stored patterns__
>
> Before running recall, we display a subset of the stored patterns to confirm the memory matrix was constructed correctly. Each column of $\mathbf{X}\in\{0,1\}^{n_{\text{pixels}}\times N}$ is reshaped to a $28\times 28$ image using the `decode(...)` function.

Let's visualize the first `number_of_show::Int` stored patterns.

```julia
let
    # initialize -
    p_list = []; #create  empty list

    # build the heatmaps -
    for k in 1:number_of_show
        img = decode(X[:, k]); #decodes the digits
        p = heatmap(img, c=:grays, colorbar=false, aspect_ratio=1, #creates visuals
                    title="Pattern $k", axis=false, ticks=false); #image characteristics
        push!(p_list, p); #add plots to list
    end

    # return -
    plot(p_list..., layout=(1, number_of_show), size=(660, 130)) #organize plot
end
```

> __Hopfield weight matrix (Hebbian rule)__
>
> The Hopfield weight matrix is $\mathbf{W} = \frac{1}{N} \mathbf{X}_{\pm 1} \mathbf{X}_{\pm 1}^{\top} \in \mathbb{R}^{n_{pixels} \times n_{pixels}}$, where $N$ is the number of stored patterns. The diagonal is zeroed to prevent trivial self-reinforcement. Recall dynamics update the state as $\mathbf{v}^{(t+1)} = \text{sign}(\mathbf{W} \mathbf{v}^{(t)})$ until convergence or for `number_of_hopfield_steps` iterations.

Let's compute the Hopfield weight matrix and store it in `W_hopfield::Array{Float64,2}`.

```julia
W_hopfield = let

    # initialize -
    N = size(X_pm1, 2); # number of stored patterns

    # TODO: compute the Hebbian weight matrix W = X_pm1 * X_pm1' / N
    # then zero the diagonal
    W = (X_pm1 * X_pm1') ./ N;
    W[diagind(W)] .= 0.0; # zero diagonal (no self-connections)

    # return -
    W
end;
```

> __Recall from a corrupted input__
>
> We select a held-out test image (not used to build the memory matrix), corrupt it by flipping `corruption_frac` $= 30\%$ of its bits using the `corrupt(...)` function, and run Hopfield dynamics for `number_of_hopfield_steps::Int` $= 20$ iterations. The Tanimoto similarity between the recalled pattern and the original measures recall quality on a scale from $0$ (no overlap) to $1$ (identical).

Let's run the Hopfield recall experiment and store the results in `v_original::Vector{Int64}` (the ground-truth held-out pattern), `v_corrupted::Vector{Int64}` (the noisy input with flipped bits), and `v_recalled::Vector{Int64}` (the Hopfield network's reconstruction).

```julia
v_original, v_corrupted, v_recalled = let

    # initialize - grab a held-out image (not in the memory matrix)
    k = 2; # which held-out image to test (k ≥ 1, offset from last stored pattern)
    img = digits_dict[digit_for_experiment][:, :, number_of_examples + k];
    v = Float64.(img)[:];           # flatten to 784-element vector
    b = (v .> 0.5);                 # threshold to binary {0,1}
    v_original = Int64.(2 .* b .- 1); # ground-truth pattern in {-1,+1}

    # corrupt the pattern - flip corruption_frac of the bits
    v_corrupted = corrupt(v_original, corruption_frac; seed=2);

    # TODO: run Hopfield dynamics for number_of_hopfield_steps steps
    # v_state <- sign(W_hopfield * v_state) at each step
    # break ties by setting 0 values to 1
    v_state = copy(v_corrupted); # placeholder - replace with dynamics
    for _ in 1:number_of_hopfield_steps
        v_state = sign.(W_hopfield * v_state); #compute v state
        v_state[v_state .== 0] .= 1; # break ties toward +1
    end
    println("Corruption fraction           : $(corruption_frac)")
    println("Tanimoto similarity (Hopfield recall): $(round(tanimoto(to_binary(v_state), to_binary(v_original)), digits=3))")

    # return -
    (v_original, v_corrupted, v_state)
end;
```

## Task 2: Train a Small RBM
The Hopfield network encodes patterns through a deterministic outer-product rule and retrieves them by energy minimization. This approach is limited by a fixed storage capacity and produces hard binary outputs with no measure of uncertainty. A [restricted Boltzmann machine (RBM)](https://en.wikipedia.org/wiki/Restricted_Boltzmann_machine) takes a different approach: it learns a probability distribution over binary patterns from training data and reconstructs inputs via block Gibbs sampling. The bipartite structure of the RBM (no within-layer connections) enables efficient parallel updates of all hidden or all visible units in a single step.

We train a small RBM with `number_of_visible::Int` $= 784$ visible units and `number_of_hidden_small::Int` $= 128$ hidden units on `number_of_training_examples::Int` $= 50$ images of digit `digit_for_experiment::Int` using contrastive divergence (CD-1). After training, we reconstruct the same corrupted held-out digit from Task 1 and compare recall quality.

> __Training data preparation__
>
> We load `number_of_training_examples` images of `digit_for_experiment` from the MNIST dataset and convert them to a $\{-1, +1\}$ integer matrix of shape $(n_{pixels} \times n_{train})$. A uniform `Categorical` distribution over the training indices serves as the data sampling distribution for the `learn(...)` function.

Let's prepare the training data and store the results in `X_train_pm1::Array{Int64,2}` (the training pattern matrix in $\{-1,+1\}$ encoding) and `p_train::Categorical` (a uniform distribution over training indices for mini-batch sampling).

```julia
X_train_pm1, p_train = let

    # initialize -
    X_train_pm1 = zeros(Int64, number_of_pixels, number_of_training_examples); # training data matrix (pixels × examples) in {-1,+1}

    # TODO: fill in X_train_pm1
    # for each i in 1:number_of_training_examples
    #   get the i-th image of digit_for_experiment, threshold at 0.5, convert to {-1,+1}
    for i in 1:number_of_training_examples
        img = digits_dict[digit_for_experiment][:, :, i]; # raw grayscale image (28 × 28)
        v = Float64.(img)[:];       # flatten to 784-element vector
        b = (v .> 0.5);             # threshold to binary {0,1}
        X_train_pm1[:, i] = Int64.(2 .* b .- 1); # convert {0,1} → {-1,+1}
    end
    p_train = Categorical(number_of_training_examples);

    println("Training data: $(size(X_train_pm1, 1)) pixels x $(size(X_train_pm1, 2)) examples")

    # return -
    (X_train_pm1, p_train)
end;
```

> __Contrastive divergence training (CD-1)__
>
> We initialize the RBM with small random weights and zero biases, then train for `number_of_epochs` epochs. Each epoch runs `number_of_updates_per_epoch` mini-batch updates using the `learn(...)` function with `T=2` Gibbs steps (CD-1, since the package counts the initial state as step 1). After each epoch, we compute the mean reconstruction error (fraction of bits incorrectly reconstructed) on a 5-image probe set.

Let's train the small RBM and store the results in `rbm_small::MyRestrictedBoltzmannMachineModel` (the trained model with learned weights and biases) and `reconstruction_errors::Vector{Float64}` (the per-epoch mean fraction of mismatched pixels on the probe set).

```julia
rbm_small, reconstruction_errors = let

    # initialize - build W (diag = 0)
    W_init = 0.01 * randn(number_of_visible, number_of_hidden_small);
    W_init[diagind(W_init)] .= 0.0; # zero diagonal (no self-connections)
    b_init = zeros(number_of_hidden_small); #creates hidden vectors
    a_init = zeros(number_of_visible); #creates visible vectors

    # build the model -
    rbm = build(MyRestrictedBoltzmannMachineModel, (
        W = W_init, # weights
        b = b_init, # hidden biases
        a = a_init  # visible biases
    ));

    # setup for reconstruction error tracking -
    number_of_probe = min(5, number_of_training_examples); #minimum of 5 training set
    probe_idx = 1:number_of_probe; #ranges of indices
    reconstruction_errors = Float64[]; #empty array to store reconstruction errors

    # training loop -
    for epoch ∈ 1:number_of_epochs
        
        # run one epoch of CD-1 training. See src/Compute.jl for implementation details.
        rbm = learn(rbm, X_train_pm1, p_train; #call training for rbm
            maxnumberofiterations = number_of_updates_per_epoch, # max weight updates per epoch
            T         = 2,              # Gibbs steps (T=2 → CD-1)
            β         = β_train,        # inverse temperature
            batchsize = batchsize_train, # mini-batch size
            η         = η_train,        # learning rate
            tol       = 1e-10,          # convergence tolerance (effectively disabled)
            verbose   = false);         # suppress per-iteration output

        # reconstruction error on probe set -
        total_error = 0.0;
        for i ∈ probe_idx
            v0 = X_train_pm1[:, i]; #store training in v0
            (V_rec, _) = sample(rbm, v0; T=2, β=β_train); # one Gibbs step: data → hidden → reconstructed visible
            total_error += mean(v0 .!= V_rec[:, end]); # fraction of bits that differ between original and reconstruction
        end
        push!(reconstruction_errors, total_error / number_of_probe); #add reconstruction errors

        # every 5 epochs, print the current epoch and reconstruction error for monitoring -
        epoch % 5 == 0 && println("Epoch $(lpad(epoch,3)) | recon error = $(round(reconstruction_errors[end], digits=4))");
    end

    # return -
    (rbm, reconstruction_errors)
end;
```

> __Training convergence__
>
> The reconstruction error (fraction of incorrectly reconstructed bits on a probe set) should decrease or plateau over training epochs as the RBM learns the co-occurrence structure of the digit patterns. You can adjust the number of epochs and updates per epoch to see how it affects convergence (if you're curious about what happens as we change these parameters).

Let's plot the reconstruction error over epochs.

```julia
let
    plot(1:number_of_epochs, reconstruction_errors, #create plot
         xlabel = "Training Epoch", #x-axis
         ylabel = "Reconstruction Error (bit fraction)", #y-axis
         title  = "Small RBM Training Convergence ($(number_of_visible) -> $(number_of_hidden_small))", #title
         label  = "Recon. Error",
         lw = 2, color = :blue, marker = :circle, size = (650, 380)) #colors
end
```

We now have a (semi) trained RBM. Let's see how well it can recall the original digit from the corrupted input.

> __Small RBM recall__
>
> We apply the same corruption used in Task 1 to the held-out digit and reconstruct it using `number_of_gibbs_recall::Int` $= 10$ steps of block Gibbs sampling with the trained small RBM. This tests whether the RBM learned a distribution that can denoise corrupted inputs.

Let's run the small RBM recall experiment and store the reconstructed pattern in `v_rbm_recalled::Vector{Int64}` (the final visible state after Gibbs sampling, in $\{-1,+1\}$ encoding).

```julia
v_rbm_recalled = let

    # initialize - apply the same corruption as Task 1
    v0_pm1 = corrupt(v_original, corruption_frac; seed=2); # corrupted input in {-1,+1}

    # run Gibbs sampling for recall -
    (V_rec, _) = sample(rbm_small, v0_pm1; T=number_of_gibbs_recall, β=β_recall); # V_rec is (pixels × T) visible trajectory
    v_rec = V_rec[:, end]; # final reconstructed pattern (last column of the chain)

    println("Tanimoto similarity (small RBM vs original): $(round(tanimoto(to_binary(v_rec), to_binary(v_original)), digits=3))")

    # return -
    v_rec
end;
```

```julia
let
    # initialize -
    imgs = [decode(v_original), decode(v_corrupted), decode(v_rbm_recalled)]; #three images: original, corrupted, recalled 
    titles = ["Original", #title
              "Δ ($(round(Int, corruption_frac*100))%)",
              "Small RBM"];

    # build the heatmaps -
    p_list = [heatmap(img, c=:grays, colorbar=false, aspect_ratio=1, #color and ratio
                      title=t, titlefontsize=9, axis=false, ticks=false) #title
              for (img, t) in zip(imgs, titles)];

    # return -
    plot(p_list..., layout=(1, 3), size=(540, 180)) #plot layout and size
end
```

> __Pretrained RBM recall__

> __Tanimoto similarity vs. corruption fraction__
>
> We sweep corruption fractions from $10\%$ to $80\%$ and run `number_of_trials::Int` $= 10$ recall trials at each level. For each trial, we corrupt a held-out digit, reconstruct it with each model, and compute the Tanimoto similarity against the original image. The sweep reveals how each model's recall quality degrades as the input becomes increasingly corrupted.

Let's compute the sweep and store the results in `accuracy_pretrained::Vector{Float64}` (mean Tanimoto similarity at each corruption level for the pretrained RBM), `accuracy_small::Vector{Float64}` (same for the small RBM), and `cf_range::Vector{Float64}` (the corruption fractions tested, from $0.1$ to $0.8$).

```julia
accuracy_pretrained, accuracy_small, cf_range = let

    # initialize -
    cf_range = [0.10, 0.20, 0.30, 0.40, 0.50, 0.60, 0.70, 0.80]; # corruption fractions to sweep
    number_of_trials = 10;  # recall trials per corruption level (averaged for robustness)
    number_of_total = size(digits_dict[digit_for_experiment], 3); # total available images for this digit
    test_start = number_of_examples + 2; # first test image index (skip stored patterns + the single held-out image)
    accuracy_pretrained = Float64[]; # mean Tanimoto similarity per corruption level (pretrained RBM)
    accuracy_small = Float64[];      # mean Tanimoto similarity per corruption level (small RBM)

    # sweep over corruption fractions -
    for cf ∈ cf_range
        total_pretrained = 0.0; # accumulator for Tanimoto scores (pretrained)
        total_small = 0.0;      # accumulator for Tanimoto scores (small)

        for trial ∈ 1:number_of_trials
            
            # initialize - grab a test image and corrupt it
            idx = min(test_start + trial - 1, number_of_total); # test image index (clamped to available images)
            img = digits_dict[digit_for_experiment][:, :, idx];
            v = Float64.(img)[:];           # flatten to 784-element vector
            b = (v .> 0.5);                 # threshold to binary {0,1}
            v_original = Int64.(2 .* b .- 1); # ground-truth pattern in {-1,+1}
            v_corrupted = corrupt(v_original, cf; seed=trial); # corrupted input (different seed per trial)

            # recall with pretrained RBM -
            (V_pt, _) = sample(rbm_pretrained, v_corrupted;
                            T=number_of_gibbs_recall, β=β_recall); # V_pt is (pixels × T) visible trajectory
            total_pretrained += tanimoto(to_binary(V_pt[:, end]), to_binary(v_original)); #total pretrained

            # recall with small RBM -
            (V_sm, _) = sample(rbm_small, v_corrupted; #sample for V sm
                            T=number_of_gibbs_recall, β=β_recall); # V_sm is (pixels × T) visible trajectory
            total_small += tanimoto(to_binary(V_sm[:, end]), to_binary(v_original)); #total small
        end

        push!(accuracy_pretrained, total_pretrained / number_of_trials); # average over trials
        push!(accuracy_small, total_small / number_of_trials);           # average over trials
    end

    # return -
    (accuracy_pretrained, accuracy_small, cf_range)
end;
```

```julia
let
    # plot accuracy vs corruption fraction for both models
    plot(cf_range, accuracy_pretrained,
        xlabel  = "Corruption Fraction", #x-axis
        ylabel  = "Tanimoto Similarity", #y-axis
        title   = "Recall Accuracy vs. Corruption Fraction (digit $(digit_for_experiment))", #title
        label   = "Pretrained RBM (784 -> 512)", #label
        lw=2, marker=:circle, color=:steelblue, size=(650, 380)); #plot characteristics
    
    plot!(cf_range, accuracy_small,
         label   = "Small RBM (784 -> $(number_of_hidden_small))", #plot label
         lw=2, marker=:square, color=:orange, linestyle=:dash); #marker color
end
```

## Summary
This problem set applied the classical Hopfield network and restricted Boltzmann machines to MNIST digit recall, comparing how deterministic energy minimization and probabilistic generative modeling handle pattern reconstruction from corrupted inputs.

> __Key Takeaways__
>
> * **Hopfield network capacity:** The Hopfield network stores patterns in a Hebbian weight matrix and retrieves them via energy minimization. Storage capacity is bounded by approximately $0.138 \times n_{\text{pixels}}$ patterns, which constrains reliable recall to a small number of stored digits relative to the full MNIST library.
> * **RBM as a generative memory:** An RBM learns a probability distribution over binary pixel patterns and reconstructs inputs via block Gibbs sampling. Contrastive divergence training allows the RBM to generalize beyond exact stored patterns to nearby corrupted inputs.
> * **Scale improves recall:** A pretrained RBM with more hidden units and training data maintains higher Tanimoto similarity across a wider range of corruption fractions than a small locally-trained model. Increased capacity enables the model to capture more complex co-occurrence patterns in the digit images.

The specialization-generalization tradeoff observed in Task 3 motivates architectures that balance model capacity with training data breadth for robust pattern recall.
